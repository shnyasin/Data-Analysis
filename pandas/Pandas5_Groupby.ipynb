# .groupby() metodu

```python
# .groupby(sütun ismi) bu kullanım ile bir sütundaki değerlere göre gruplama yapılabilir.
# Devamında grupladıktan sonra bir aggregate function'ına bakmalıyız
# groupby() yaptıktan sonra aşağıdaki fonksiyonların birini ya da birkaçını kullanmak zorundayız. Yoksa bir anlamı kalmıyor gruplamanın !!!
```

Common Options:

    mean(): Compute mean of groups
    sum(): Compute sum of group values
    size(): Compute group sizes
    count(): Compute count of group
    std(): Standard deviation of groups
    var(): Compute variance of groups
    sem(): Standard error of the mean of groups
    describe(): Generates descriptive statistics
    first(): Compute first of group values
    last(): Compute last of group values
    nth() : Take nth value, or a subset if n is a list
    min(): Compute min of group values
    max(): Compute max of group values
    
sum():
Gruplardaki değerlerin toplamını verir.

size():
Her gruptaki eleman sayısını verir (NaN değerleri de sayar).

count():
Her gruptaki geçerli (NaN olmayan) değerlerin sayısını verir.

std():
Gruplardaki değerlerin standart sapmasını hesaplar.

var():
Gruplardaki değerlerin varyansını hesaplar.

sem():
Gruplardaki ortalamanın standart hatasını verir.

describe():
Gruplar için temel istatistikleri (count, mean, std, min, 25%, 50%, 75%, max) gösterir.

first():
Her grubun ilk değerini verir.

last():
Her grubun son değerini verir.

nth(n):
Her grubun n. sırasındaki değeri verir (örneğin nth(2) → 3. eleman).
Liste olarak da kullanılabilir: nth([0, 1]) → ilk iki elemanı verir.

Full List at the Online Documentation: https://pandas.pydata.org/docs/reference/groupby.html
