# Missing Valuelarla Yapılabilecek 3 şey vardır

```python
# 1) Dokunmadan bırakabiliriz.
# 2) Drop edebiliriz(Satır veya sütun bazlı)(En son başvurulan yöntem)
# 3) Mantıklı bir şekilde doldurabiliriz.
```

```python
df
```

```python
# 1.Yol : Yapılacak hiçbir şey yok. Bu şekilde işlemlerine devam et.
```

```python
# 2.Yol :
```

```python
df.dropna()   #nan olan her şeyi drop eder.(İstediğimiz bir şey değil bu)
```

```python
df.dropna(how = 'any') # Herhangi bir veri nan ise o satırı düşür. Üstekinin defaultu how = 'any'dir.
```

```python
df.dropna(how = 'all') # Bütün verileri nan ise o satırı düşür.
```

```python
df.dropna(thresh= 1) # 1 tane bile nan olmayan veri varsa o satırı tut. thresh eşik değeri demek
```

```python
df
```

```python
df.dropna(thresh = 4)
```

```python
df.dropna(thresh = 5)
```

```python
df.head()
```

```python
df.dropna(axis=1)  #herhangi bir sütunda 1 tane bile nan veri varsa o sütunu düşürür.
```

```python
df.dropna(axis = 1 , thresh = 4)  #pre_movie_score ve post_movie_score'da 3 nan olmayan değer olduğu için bu 2 sütunu düşürür.
```

```python
# 3.Yol : Fill the data(Datayı doldur.)
```

```python
df.head()
```

```python
df.dropna(how = 'all' , inplace = True)  #tüm verileri nan olan verileri tamamen silme. Yani 1. index tamamen silinir.
```

```python
df.head()
```

```python
df['pre_movie_score'].fillna(9.5)  #fillna nan'ları doldurma anlamına geliyor. Burada 9.5 ile doldurduk. Manuel doldurma.
```

```python
df['pre_movie_score'].fillna(df['pre_movie_score'].mean())  #pre_movie_scor'daki nan'ları pre_movie_score'un ortalamasıyla doldurduk. Dinamik doldurma.
```
