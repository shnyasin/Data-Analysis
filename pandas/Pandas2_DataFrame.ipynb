```python
my_idx = ['TR' , 'BR' , 'USA' , 'RU']
my_cols = ['Jan' , 'Feb' , 'Mar']
my_data = np.random.randint(0,101, size= (4,3))
```

# DataFrame'de Sorgu İşlemleri

```python
df = pd.DataFrame({'Year' : [1776,1867,1821] , 'Pop' : [328,38,126] , 'GDP' : [20.5,1.7,1.22]} , index = ['USA' , 'CANADA' , 'MEXICO']) # Dict vererek DF oluşturma.
df
```

```python
# Hangi ülkelerin 50 milyon'dan daha fazla popülasyonu var ?
```

```python
df['Pop']
```

```python
df['Pop'] > 50  #tek df'te true false olarak cevap verir
```

```python
df[df['Pop'] > 50] #çift df'te ise içindeki sorguya göre bu sorguya uyanları yazdırır
```

```python
# Hangi ülkeler 1800 yılından sonra kuruldu ?

df[df['Year'] > 1800]
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
# Soru : Ödenen ücreti 40'dan büyük olan satırları getirin.
```

```python
df[df['total_bill'] > 40]
```

```python
# Soru : Cinsiyetleri Erkek olan satırları getirin.
```

```python
df[df['sex'] == 'Male']
```

```python
# Soru : Cinsiyetleri Kadın olan VE bahşiş miktarı 5'den büyük olan satırları getirin.
```

```python
df[(df['sex'] == 'Female') & (df['tip'] > 5) ] # & ve sembolü ayrıca sorguları parantezde göstermek ZORUNDAYIZ
```

```python
# Soru : Günü Pazar(Sun) olan VE ödenen miktar 30'dan büyük olan satırları getirin.
```

```python
df[(df['day'] == 'Sun') & (df['total_bill'] > 30)]
```

```python
# Soru : Cumartesi(Sat) VEYA Pazar(Sun) günü olan satırları getirin.
```

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun')] # | işaret veya işareti
```

```python
# Soru : Cumartesi(Sat) veya Pazar(Sun) veya Perşembe(Thur) günü olan satırları getirin.
```

```python
df[(df['day'] == 'Sat') | (df['day'] == 'Sun') | (df['day'] == 'Thur')]
```

```python
df[df['day'].isin(['Sat' , 'Sun' , 'Thur'])]  #isin fonksiyonu içinde mi anlamına gelir. Üstekinin daha kolay hali.
```

```python
df[df['day'] != 'Sun']
```

```python
df[~(df['day'] == 'Sun')]   # ~ ile bunlar dışındakileri getir diyebiliriz. Sun olmayan günleri yaz. Üsttekinin başka yolu
```
